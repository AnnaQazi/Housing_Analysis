```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV
df=pd.read_csv(r"C:\Users\pds\Documents\Housing.csv")
```

```python
# Show First 5 rows 
df.head()
```

```python
# Statistical summary
df.describe()
```

```python
# Check overall dataset info: rows, columns, datatypes, missing values
df.info()
```

```python
# Columns to analyze statistics
cols = ['price', 'area', 'bedrooms']

for col in cols:
    print(f"\n--- {col} ---")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode()[0])
    print("Variance:", df[col].var())
    print("Standard Deviation:", df[col].std())
    print("Skewness:", df[col].skew())
    print("Kurtosis:", df[col].kurt())
    print("Min:", df[col].min())
    print("Max:", df[col].max())
```

```python
plt.figure(figsize=(14,5))

# Price Histogram
plt.subplot(1, 2, 1)  # 1 row, 2 cols, plot 1
sns.histplot(df['price'], kde=True, color='skyblue')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')

# Area Histogram
plt.subplot(1, 2, 2)  # 1 row, 2 cols, plot 2
sns.histplot(df['area'], kde=True, color='salmon')
plt.title('Area Distribution')
plt.xlabel('Area')
plt.ylabel('Count')

plt.tight_layout()# Makes sure labels and titles don’t overlap
plt.show()
```

```python
# Scatter Plot: Price vs Area
plt.figure(figsize=(8,5))
sns.scatterplot(x='area', y='price', data=df)
plt.title('Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

### Data Story

I explored the housing dataset to understand the price, area, and bedroom distributions.
The average house price is around 4,766,729, but prices range widely from 1,750,000 up to 13,300,000. The high standard deviation (~1,870,439) and positive skew (1.21) show that while most homes cluster around the median price of 4,340,000, a few very expensive properties are pulling the average up.
Looking at the area, homes average about 5150 units, ranging from 1650 to 16,200 . Area also shows a positive skew (1.32), indicating many smaller to mid-sized houses but a few really large ones  Bedroom counts are more consistent, averaging just under 3 with little variation.

Overall, the data tells the housing market is pretty mixed. There are many average sized, average priced homes, but a noticeable number of bigger and pricier houses. Also, the spread and right skew in price and area suggests there are some luxury homes that are much bigger and costlier than average those are the outliers pushing the market up. It shows that while area has a strong impact on price, it’s not the only factor. The data hints that houses with more area generally cost more, but some small houses are still expensive probably due to other luxury features.
